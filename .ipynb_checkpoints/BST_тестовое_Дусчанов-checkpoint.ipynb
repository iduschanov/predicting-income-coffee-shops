### 1. Очистка данных 

#### 1.1 Пропущеные значения.

```python
# Проверим сколько у нас столбцов с категориальными данными
data.select_dtypes(include=['object']).shape
```

```python
data.head()
```

```python
# От 88 до 99 строки в датасете начинается тестовая часть поэтому ее вырезаем и сохраняем в тесте.
test_df = data.iloc[88:99]
data.drop(index=data.iloc[88:99].index.tolist(), inplace=True)
# Оставшиеся данные в data записываем в train_df
train_df = data.copy()
```

```python
test_persentage = test_df.drop("target", axis=1).isnull().sum().sort_values(ascending=False)

# Удаляем столбцы, где все значения в тесте пропуски.
train_df = train_df.drop(test_persentage[test_persentage == len(test_df)].index, 1)
test_df = test_df.drop(test_persentage[test_persentage == len(test_df)].index, 1)

train_df.shape, test_df.shape
```

```python
test_std = test_df.drop("target", axis=1).std().sort_values(ascending=False)

test_same_not_null_values = test_persentage[
    (test_persentage.index.isin(test_std[test_std == 0].index)
    ) & (test_persentage == 0)
].index

# Удаляем столбцы, где в тестовых данных нет пропусков и все знаечния одинаковы.
train_df = train_df.drop(test_same_not_null_values, 1)
test_df = test_df.drop(test_same_not_null_values, 1)

train_df.shape, test_df.shape
```

```python
# Выводим размеры train/test датасетов
train_df.shape, test_df.shape
```

```python
# Смотрим сколько пропущенных данных. 
# Total - это общее кол-во пропусков, Percent - это процент пропущенных данных.
total = train_df.isnull().sum().sort_values(ascending=False)
percent = (train_df.isnull().sum()/train_df.isnull().count()).sort_values(ascending=False)
missing_data = pd.concat([total, percent], axis=1, keys=['Total', 'Percent'])
missing_data.head(10)
```

```python
# Удаляем столбцы, где пропусков больше 40 значений.
train_df = train_df.drop((missing_data[missing_data['Total'] > 40]).index,1)
test_df = test_df.drop((missing_data[missing_data['Total'] > 40]).index,1)
# тут можно было использовать kNN для заполнения пропусков, но столбцов с пропущенными данными очень много.
# Запоняем нулями оставшиеся пропуски. 
# При просмотре данных некоторые столбцы имеют бинарные значения, но пи это где 0 у нас пропуски.
train_df = train_df.apply(lambda x:x.fillna(0)) #x.value_counts().index[0]
test_df = test_df.apply(lambda x:x.fillna(0)) #x.value_counts().index[0]
# Проверм сколько пропусков осталось.
train_df.isnull().sum().max(), test_df.isnull().sum().max()
```

#### 1.2 Удаляем коррелирующие признаки

```python
# Удаляем значения, которые имеют корреляцию по модулю больше 0.90
corr = abs(train_df.corr()) 
upper_tri = corr.where(np.triu(np.ones(corr.shape),k=1).astype(np.bool))
to_drop = [column for column in upper_tri.columns if any(upper_tri[column] > 0.90)]

train_df.drop(to_drop, axis=1, inplace=True)
test_df.drop(to_drop, axis=1, inplace=True)
```

```python
len(to_drop), train_df.shape, test_df.shape
```

#### 1.3 Фильтрация данных

```python
# Удаляем константные колонки с низкой дисперсией и дубликаты колонок, если они есть.
threshold_n=0.95
sel = VarianceThreshold(threshold=0.95)
sel_var=sel.fit_transform(train_df)
col = train_df.columns[sel.get_support(indices=True)].tolist()
col.append('target') # Target нам нужен поэтому ее оставляем.
train_df = train_df[col]
test_df = test_df[col]
```

```python
train_df.shape, test_df.shape
```

```python
def model_importances(estimator, data):
  '''DataFrame с важными фичами''' 
  feat_dict= {}
  for col, val in sorted(zip(data.columns, estimator.feature_importances_),key=lambda x:x[1],reverse=True):
    feat_dict[col]=val 
  return pd.DataFrame({'Feature':feat_dict.keys(),'Importance':feat_dict.values()})
```

### 3. Feature engineering 

#### 3.1 Polynomial Features

При удалении небольших выбросов в данных, набор данных для теста сильно уменьшается. Следовательно, качество моделей падает.

У нас данные не имеют линейную связь с целевой переменно поэтому добавим полиномиальные фичи 2 степени топ 20 фич по важности признаков RF. 

```python
from sklearn.preprocessing import PolynomialFeatures

poly = PolynomialFeatures(2, include_bias=False)
output_train = poly.fit_transform(train_df[feat_df['Feature'].iloc[:20].values])
output_test = poly.transform(test_df[feat_df['Feature'].iloc[:20].values])

train_poly_df = pd.DataFrame(output_train, columns = 
                             poly.get_feature_names(train_df[feat_df['Feature'].iloc[:20].values].columns))
test_poly_df = pd.DataFrame(output_test, columns = 
                            poly.get_feature_names(train_df[feat_df['Feature'].iloc[:20].values].columns))
```

```python
# Из-за того, что LGBM ругается, что символ ^ не относиться к ASCII символам, заменяем его на слово power.
train_poly_df = train_poly_df.rename(columns = lambda x: x.replace('^', ' power '))
test_poly_df = test_poly_df.rename(columns = lambda x: x.replace('^', ' power '))
print(train_poly_df.shape, test_poly_df.shape)
```

```python
used_columns = set(train_poly_df.columns) - set(train_df.columns)
train_poly_df[used_columns].shape, train_df.shape
```

```python
train_df.shape, test_df.shape
```

```python
# Добавляем полиномиальные фичи к основным данным.
train_df = pd.concat(
    [train_df.reset_index(drop=True),
    train_poly_df[used_columns].reset_index(drop=True)],
    axis=1
    )

test_df = pd.concat(
    [test_df.reset_index(drop=True),
    test_poly_df[used_columns].reset_index(drop=True)],
    axis=1
    )
train_df.shape, test_df.shape
```

#### 4.3 RF

```python
def objective(trial, X, y, N_FOLDS, RANDOM_STATE):

    param_grid_rf = {
        "n_estimators": trial.suggest_int("n_estimators", 31, 128, step=10),
        "max_depth": trial.suggest_int("max_depth", 3, 20, step=2),
        "min_samples_split": trial.suggest_int("min_samples_split", 3, 10, step=2)
        }
        
    cv = KFold(n_splits=N_FOLDS, shuffle=True, random_state=RANDOM_STATE)

    cv_predicts = np.empty(N_FOLDS)
    for idx, (train_idx, test_idx) in enumerate(cv.split(X, y)):
        X_train, X_test = X.iloc[train_idx], X.iloc[test_idx]
        y_train, y_test = y.iloc[train_idx], y.iloc[test_idx]

        model = RandomForestRegressor(n_jobs=-1, random_state=RANDOM_STATE,
                                      **param_grid_rf, criterion="squared_error")
        model.fit(X_train,
                  y_train
                  )
        preds = model.predict(X_test)        
        cv_predicts[idx] = mean_absolute_percentage_error(y_test, preds)

    return np.mean(cv_predicts)
```

```python
study_rf = optuna.create_study(direction="minimize", study_name="RF")
func = lambda trial: objective(trial, x_train, y_train, N_FOLDS, RANDOM_STATE)
# n_trials - кол-во итераций
study_rf.optimize(func, n_trials=20, show_progress_bar=True)
```

```python
model_rf = RandomForestRegressor(n_jobs=-1, random_state=RANDOM_STATE, **study_rf.best_params).fit(x_train, y_train)
mean_absolute_percentage_error(y_test, model_rf.predict(x_test)) 
```

#### 4.4 LGBM

```python
def objective(trial, X, y, N_FOLDS, RANDOM_STATE):

    param_grid_lg = {
        "n_estimators": trial.suggest_int("n_estimators", 31, 64, step=8),
        "learning_rate": trial.suggest_float("learning_rate", 0.001, 0.05, step=2),
        "num_leaves": trial.suggest_int("num_leaves", 20, 300, step=10),
        "max_depth": trial.suggest_int("max_depth", 3, 10, step=2),
        "min_child_samples": trial.suggest_int("min_child_samples", 5, 200, step=10),
        }

    cv = KFold(n_splits=N_FOLDS, shuffle=True, random_state=RANDOM_STATE)

    cv_predicts = np.empty(N_FOLDS)
    for idx, (train_idx, test_idx) in enumerate(cv.split(X, y)):
        X_train, X_test = X.iloc[train_idx], X.iloc[test_idx]
        y_train, y_test = y.iloc[train_idx], y.iloc[test_idx]

        pruning_callback = optuna.integration.LightGBMPruningCallback(
            trial, "l2")
        model_x = LGBMRegressor(objective="regression", **param_grid_lg, random_state=RANDOM_STATE)
        model_x.fit(X_train,
                  y_train,
                  eval_set=[(X_test, y_test)],
                  eval_metric="neg_mean_absolute_percentage_error",
                  callbacks=[pruning_callback],
                  verbose=-1
                  )
        preds = model_x.predict(X_test)        
        cv_predicts[idx] = mean_absolute_percentage_error(y_test, preds)

    return np.mean(cv_predicts)
```

```python
study_lg = optuna.create_study(direction="minimize", study_name="LGB")
func = lambda trial: objective(trial, x_train, y_train, N_FOLDS, RANDOM_STATE)
# n_trials - кол-во итераций
study_lg.optimize(func, n_trials=20, show_progress_bar=True)
```

```python
model_lg = LGBMRegressor(objective="regression", **study_lg.best_params, 
                      random_state=RANDOM_STATE).fit(x_train, y_train)
mean_absolute_percentage_error(y_test, model_lg.predict(x_test))
```

#### 4.5 CatBoost

```python
def objective(trial, X, y, N_FOLDS, RANDOM_STATE):

    param_grid = {
        "iterations": trial.suggest_categorical("iterations", [400, 600, 700, 900, 1100, 1300,]),
        "learning_rate": trial.suggest_float("learning_rate", 0.01, 0.3, step=0.01),
        "max_depth": trial.suggest_int("max_depth", 3, 10, step=2),
        "min_child_samples": trial.suggest_int("min_child_samples", 5, 200, step=10),
        }

    cv = KFold(n_splits=N_FOLDS, shuffle=True, random_state=RANDOM_STATE)

    cv_predicts = np.empty(N_FOLDS)
    for idx, (train_idx, test_idx) in enumerate(cv.split(X, y)):
        X_train, X_test = X.iloc[train_idx], X.iloc[test_idx]
        y_train, y_test = y.iloc[train_idx], y.iloc[test_idx]

        model = CatBoostRegressor(**param_grid, random_state=RANDOM_STATE, loss_function="MAPE", verbose=False)
        model.fit(X_train,
                  y_train
                  )
        preds = model.predict(X_test)        
        cv_predicts[idx] = mean_absolute_percentage_error(y_test, preds)

    return np.mean(cv_predicts)
```

```python
study_cat = optuna.create_study(direction="minimize", study_name="CT")
func = lambda trial: objective(trial, x_train, y_train, N_FOLDS, RANDOM_STATE)
# n_trials - кол-во итераций
study_cat.optimize(func, n_trials=2, show_progress_bar=True)
```

```python
model_cat = CatBoostRegressor(**study_cat.best_params, verbose=False).fit(x_train, y_train)
mean_absolute_percentage_error(y_test, model_cat.predict(x_test))
```

#### 4.6 GradientBoosting

```python
def objective(trial, X, y, N_FOLDS, RANDOM_STATE):

    param_grid = {
        "n_estimators": trial.suggest_int("n_estimators", 31, 255, step=8),
        "learning_rate": trial.suggest_float("learning_rate", 0.01, 0.3),
        "min_samples_leaf": trial.suggest_int("min_samples_leaf", 20, 300, step=10),
        "max_depth": trial.suggest_int("max_depth", 3, 10, step=2)
        }

    cv = KFold(n_splits=N_FOLDS, shuffle=True, random_state=RANDOM_STATE)

    cv_predicts = np.empty(N_FOLDS)
    for idx, (train_idx, test_idx) in enumerate(cv.split(X, y)):
        X_train, X_test = X.iloc[train_idx], X.iloc[test_idx]
        y_train, y_test = y.iloc[train_idx], y.iloc[test_idx]

        model = GradientBoostingRegressor(**param_grid, random_state=RANDOM_STATE,
                                          loss="squared_error")
        model.fit(X_train,
                  y_train
                  )
        preds = model.predict(X_test)        
        cv_predicts[idx] = mean_absolute_percentage_error(y_test, preds)

    return np.mean(cv_predicts)
```

```python
study_gbr = optuna.create_study(direction="minimize", study_name="GBR")
func = lambda trial: objective(trial, x_train, y_train, N_FOLDS, RANDOM_STATE)
# n_trials - кол-во итераций
study_gbr.optimize(func, n_trials=20, show_progress_bar=True)
```

```python
model_gbr = GradientBoostingRegressor(**study_gbr.best_params, 
                                  loss="squared_error").fit(x_train, y_train)
mean_absolute_percentage_error(y_test, model_gbr.predict(x_test)) 
```

### 5. Cross-validation & learning_curve

```python
def plot_learning_curve(estimator, title, X, y, ylim=None, cv=None,
                        n_jobs=None, train_sizes=np.linspace(.1, 1.0, 5)):
    """
    График кривой обучения для test and training learning curve.

    Parameters
    ----------
    estimator : object type модель для обучения.

    X : array-like, обучающий набор данных.

    y : array-like, целевая переменная.

    cv : int, генератор cross-validation.

    """
    plt.figure()
    plt.title(title)
    if ylim is not None:
        plt.ylim(*ylim)
    plt.xlabel("Training examples")
    plt.ylabel("Score")
    
    train_sizes, train_scores, test_scores = learning_curve(
        estimator, X, y, cv=cv, n_jobs=n_jobs, train_sizes=train_sizes, scoring=mape_score)
    train_scores_mean = np.mean(train_scores, axis=1)
    train_scores_std = np.std(train_scores, axis=1)
    test_scores_mean = np.mean(test_scores, axis=1)
    test_scores_std = np.std(test_scores, axis=1)
    plt.grid()

    print('train_scores_mean: ', train_scores_mean, '\ntest_scores_mean:', test_scores_mean)

    plt.fill_between(train_sizes, train_scores_mean - train_scores_std,
                     train_scores_mean + train_scores_std, alpha=0.1,
                     color="r")
    plt.fill_between(train_sizes, test_scores_mean - test_scores_std,
                     test_scores_mean + test_scores_std, alpha=0.1, color="g")
    plt.plot(train_sizes, train_scores_mean, 'o-', color="r",
             label="Training score")
    plt.plot(train_sizes, test_scores_mean, 'o-', color="g",
             label="Cross-validation score")

    plt.legend(loc="best")
    return plt

cv = ShuffleSplit(n_splits=N_FOLDS, test_size=0.2, random_state=RANDOM_STATE)

x_train, x_test, y_train, y_test = train_test_split(train_df.drop('target', axis=1),
                                                    train_df['target'], test_size=.2, random_state=RANDOM_STATE)
```

```python
title = "Learning Curve RandomForestRegressor"
estimator = RandomForestRegressor(n_jobs=-1, random_state=RANDOM_STATE,
                          **study_rf.best_params, criterion="squared_error")
plot_learning_curve(estimator, title, x_train, y_train, cv=cv)

plt.show()
```

```python
title = "Learning Curve LGBMRegressor"
estimator = LGBMRegressor(objective="regression", **study_lg.best_params, 
                      random_state=RANDOM_STATE).fit(x_train, y_train)
plot_learning_curve(estimator, title, x_train, y_train, cv=cv)

plt.show()
```

```python
title = "Learning Curve CatBoostRegressor"
estimator = CatBoostRegressor(**study_cat.best_params, random_state=RANDOM_STATE,
                              loss_function="MAPE", verbose=False)
plot_learning_curve(estimator, title, x_train, y_train, cv=cv)

plt.show()
```

```python
title = "Learning Curve GradientBoostingRegressor"
estimator = GradientBoostingRegressor(**study_gbr.best_params, loss="squared_error",
                              random_state=RANDOM_STATE)
plot_learning_curve(estimator, title, x_train, y_train, cv=cv)

plt.show()
```

В модель добавляем линейную модель SVR, если в тесте появится новое значение среднего дохода выше или ниже, чем было в таргете в трейне, чтоб ансамбль моделей мог экстраполировать. Модели основнные на деревьях решений умееют только интерполировать, но не экстраполировать строиться на основе интерполяции.

```python
title = "Learning Curve VotingRegressor"
estimators=[('RF', RandomForestRegressor(n_jobs=-1, random_state=RANDOM_STATE, **study_rf.best_params)),
            ('LGBM', LGBMRegressor(objective="regression", **study_lg.best_params, 
                      random_state=RANDOM_STATE)), 
            ('Catboost', CatBoostRegressor(**study_cat.best_params, random_state=RANDOM_STATE,
                              loss_function="MAPE", verbose=False)), 
            ('GBR', GradientBoostingRegressor(**study_gbr.best_params, loss="squared_error",
                              random_state=RANDOM_STATE)), 
            ('SVR', SVR())]
estimator = VotingRegressor(estimators)

plot_learning_curve(estimator, title, train_df.drop('target', axis=1), train_df['target'], cv=cv)
plt.show()
```
